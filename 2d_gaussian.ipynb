```python
import numpy as np
import matplotlib.pyplot as plt
# This is similar to the code from lecture 1
# to sample from a 2D Gaussian Distribution
mean = [0,0]
cov = [[1, 1.5], [1.5, 5]]
# Fix random seed to get consistent results
np.random.seed(1024)
X = np.random.multivariate_normal(mean, cov, 1000)
fig, ax = plt.subplots(figsize=(10, 10))
# c='r', dot color is red
# s=10.0, dot size is 10
# alpha=0.3, dot opacity is 0.3
ax.scatter(X[:,0], X[:,1], c='r', s=10.0, alpha=0.3, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
# Set x/y axis limits
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
fig.show()
```

```python
# TODO: Report the estimate mean and covariance
# of the sampled points
print ("Shape of X:", X.shape)
print ("Mean of X: ", X.mean(axis=0))
print ("Covariance of X: ")
print(np.cov(X, rowvar=False))
```

```python
# TODO: Plot the histogram for the x-coordinates of X
# and y-coordinates of X respectively.
# You can use the plt.hist() function

fig, ax = plt.subplots(figsize=(10, 10))
ax.grid()
plt.hist(X[:,0], bins=30, facecolor='blue', alpha=0.5)
plt.title('Histogram of x-coordinate values', fontsize=20)
plt.xlabel('values', fontsize=14)
plt.ylabel('freq', fontsize=14)
plt.show()

fig, ax = plt.subplots(figsize=(10, 10))
ax.grid()
plt.hist(X[:,1], bins=30, facecolor='blue', alpha=0.5)
plt.title('Histogram of y-coordinate values', fontsize=20)
plt.xlabel('values', fontsize=14)
plt.ylabel('freq', fontsize=14)
plt.show()
```

```python
# TODO: Are the x-coordinates of X samples from
# some Gaussian distribution?
# If so, estimate the mean and variance.
# Do the same for the y-coordinates.

x_mean = np.mean(X[:,0])
x_var = np.var(X[:,0])
y_mean = np.mean(X[:,1])
y_var = np.var(X[:,1])

print ("Mean of gaussian distribution of x-coordinates of new X samples: % .7f" % x_mean)
print ("Variance of gaussian distribution of x-coordinates of new X samples: % .7f" % x_var)
print ("Mean of gaussian distribution of y-coordinates of new X samples: % .7f" % y_mean)
print ("Variance of gaussian distribution of y-coordinates of new X samples: % .7f" % y_var)

x_sigma = np.sqrt(x_var)
y_sigma = np.sqrt(y_var)

fig, ax = plt.subplots(figsize=(20, 10))
plt.subplot(1,2,1)
count, bins, ignored = plt.hist(X[:,0], bins=30, density=True)
plt.plot(bins, 1/(x_sigma * np.sqrt(2 * np.pi))*np.exp(- (bins-x_mean)**2 / (2*x_sigma**2)), linewidth=2, color='r')
plt.title('Histogram of x-coordinate values', fontsize=20)
plt.xlabel('values', fontsize=14)
plt.ylabel('freq density', fontsize=14)

plt.subplot(1,2,2)
count, bins, ignored = plt.hist(X[:,1], bins=30, density=True)
plt.plot(bins, 1/(y_sigma * np.sqrt(2 * np.pi))*np.exp(- (bins-y_mean)**2 / (2*y_sigma**2)), linewidth=2, color='r')
plt.title('Histogram of y-coordinate values', fontsize=20)
plt.xlabel('values', fontsize=14)
plt.ylabel('freq density', fontsize=14)


# TODO: Generate a new 2D scatter plot of 1000 points,
# such that the x-coordinates(y-coordinates) of all the points
# are samples from a 1D Gaussian distribution 
# using the estimated mean and variance based on the x-coordinates(y-coordinates) of X.

sample_x_coords = np.random.normal(x_mean, x_sigma, 1000)
sample_y_coords = np.random.normal(y_mean, y_sigma, 1000)

fig, ax = plt.subplots(figsize=(10, 10), dpi=300)
# c='r', dot color is red
# s=10.0, dot size is 10
# alpha=0.3, dot opacity is 0.3
ax.scatter(sample_x_coords, sample_y_coords, c='r', s=10.0, alpha=0.3, label="1D-Gaussian along each dimension")
ax.grid()
ax.legend(loc = 0)
# Set x/y axis limits
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
plt.show()
```

```python
# Back to the original X
fig, ax = plt.subplots(figsize=(10, 10))
# c='r', dot color is red
# s=10.0, dot size is 10
# alpha=0.3, dot opacity is 0.3
ax.scatter(X[:,0], X[:,1], c='r', s=10.0, alpha=0.3, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

# TODO: Plot a line segment with x = [-10, 10]
# and y = 3x + 1 onto the 2D-Gaussian plot.
# The np.linspace() function may be helpful.
x_points = np.linspace(start=-10, stop=10, num=200, endpoint=True)
y_points = 3*x_points + np.ones(x_points.shape)
ax.plot(x_points, y_points, label='y=3x+1')
ax.legend(loc = 0)
fig.show()
```

```python
def find_point_on_line(a, b, c, x1, y1):
    const = (-1*(a*x1 + b*y1 + c) / (a*a + b*b))  
    x = const*a + x1  
    y = const*b + y1  
    return np.array([x, y])
```

```python
# TODO: Project X onto line y=3x + 1
# and plot the projected points on the 2D space.
# You need to remove this line and assign the projected points to X_proj.
X_proj = np.copy(X)
for i in range(X_proj.shape[0]):
    X_proj[i,:] = find_point_on_line(-3, 1, -1, X[i,0], X[i,1])

    
# Here we only plot 30 points to check the correctness
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(X_proj[:30,0], X_proj[:30,1], c='b', s=10.0, alpha=1.0, label="Projected 2D-Gaussian")
ax.scatter(X[:30,0], X[:30,1], c='r', s=10.0, alpha=1.0, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
# You can also add the line from the previous plot for verification.

ax.plot(x_points, y_points, label='y=3x+1')
ax.legend(loc = 0)
fig.show()
```
